## Run fastdup

Once the extraction completes, we can run fastdup on the images.

For that let's initialize fastdup and specify the input directory which points to the folder of images.

```python
fd = fastdup.create(input_dir="images/")
```

> 🗒 **Note** - The `.create` method also has an optional `work_dir` parameter which specifies the directory to store artifacts from the run.

In other words you can run `fastdup.create(input_dir="images/", work_dir="my_work_dir/")` if you'd like to store the artifacts in a `my_work_dir`.

Now, let's run fastdup.

```python
fd.run()
```

## View Run Summary

After the run is completed, you can optionally view the summary with:

```python
fd.summary()
```

## Duplicate/Near-duplicates

One of the lowest hanging fruits in cleaning a dataset is finding and eliminating duplicates.

fastdup provides a handy way of visualizing duplicates/near-duplicates using the `duplicates_gallery` method. The `Distance` value indicates how visually similar are the image pairs in the gallery. A `Distance` of `1.0` indicates an exact copy and vice-versa.

```python
fd.vis.duplicates_gallery()
```

## Removing Duplicates

Let's first get the information about which cluster each image belongs to.

```python
connected_components_df , _ = fd.connected_components()
connected_components_df
```

Duplicates are stored in a cluster (component id). Let's group them by componenet_id

```python
duplicates_df = connected_components_df.groupby('component_id').agg({
    'filename': list,
    'count': len,
    'mean_distance': 'mean',
})
```

```python
duplicates_df
```

```python
import pandas as pd

def extract_image_duplicates(row):
    filenames = row['filename']
    image = filenames[0]
    duplicates = filenames[1:] if len(filenames) > 1 else []
    return pd.Series({'image': image, 'duplicates': duplicates})

df = duplicates_df.apply(extract_image_duplicates, axis=1)
df
```

```python
from IPython.display import HTML

def display_image(image_path):
    return f'<img src="{image_path}" width="100">'

def display_duplicates(duplicates):
    if isinstance(duplicates, list):
        return ''.join([f'<img src="{duplicate}" width="100">' for duplicate in duplicates])
    else:
        return ''

# Apply the display_image function to the 'image' column
df['image_display'] = df['image'].apply(display_image)

# Apply the display_duplicates function to the 'duplicates' column
df['duplicates_display'] = df['duplicates'].apply(display_duplicates)

# Display the DataFrame with the image columns
df = df.style.format({'image_preview': lambda x: x, 'duplicates_preview': lambda x: x})
display(df)
```

```python
duplicates_to_remove = df['duplicates'].tolist()
```

```python
duplicates_to_remove
```
